```python
import numpy as np
import scipy as sp
import pandas as pd
import seaborn as sb

def readTrainDataframe():
    result = pd.read_csv("../Bases/treino.csv")
    result.drop('Index', axis=1, inplace=True)
    return result

def readValidationDataframe():
    result = pd.read_csv("../Bases/validacao.csv")
    result.drop('Index', axis=1, inplace=True)
    return result

def readTestDataframe():
    result = pd.read_csv("../Bases/teste.csv")
    result.drop('Index', axis=1, inplace=True)
    return result
```

```python
df_train = readTrainDataframe()
df_validation = readValidationDataframe()
df_test = readTestDataframe()
df_original = pd.read_csv("../Bases/creditcard.csv")

print(f"Len of df_train:      {len(df_train):6}")
print(f"Len of df_validation: {len(df_validation):6}")
print(f"Len of df_test:       {len(df_test):6}")
print(f"Len of df_original:   {len(df_original):6}")
df_train = None
df_validation = None
df_test = None
df_original = None
```

```python
df_test = readTestDataframe()
auxColsDf = df_test.columns
colsDf = []

for i in range(len(auxColsDf)):
    if (auxColsDf[i] != 'Index'):
        colsDf.append(auxColsDf[i])

print(colsDf)
auxColsDf = None
df_test = None
```

```python
from scipy.spatial.distance import pdist, squareform
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 300


df_validation = readValidationDataframe()

# Calculate distance matrix
distance_matrix = squareform(pdist(df_validation.iloc[lambda x: x.index % 3 == 0], 'euclidean'))
df_validation = None

# Visualization distance matrix
fig, ax = plt.subplots()
fig.set_figwidth(30)
fig.set_figheight(30)
cax = ax.matshow(distance_matrix, cmap='viridis')
distance_matrix = None
# Add color bar
cbar = fig.colorbar(cax)
cbar.ax.set_ylabel('Distance', rotation=270, labelpad=15)

# Show chart
plt.title('Distance Matrix validation')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

cax = None
cbar = None
```
